```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
n_samples = 1000
n_clusters = 5
random_state = 50
X, y = make_blobs(n_samples=n_samples, centers=n_clusters, random_state=random_state)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Sample Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
kmeans.fit(X)
```

```python
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, s=50, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, alpha=0.5)
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
n_samples = 1000
n_clusters = 5
random_state = 50
X, y = make_blobs(n_samples=n_samples, centers=n_clusters, random_state=random_state)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Sample Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()


linked = linkage(X, 'ward')


plt.figure(figsize=(13, 8))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
plt.show()
```
